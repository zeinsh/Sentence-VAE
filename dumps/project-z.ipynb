```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import json
from tqdm import tqdm
%matplotlib notebook

```

```python
plt.figure()

with open('./zero/valid_E9.json') as fin:
    datastore = json.loads(fin.read())
    
z, sentences=datastore['z'][:2000], datastore['target_sents'][:2000]
# find tsne coords for 2 dimensions
tsne = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
Y = tsne.fit_transform(z)

plt.subplot(2,2,1)
plt.axis('off')

plt.title('Zero')
x_coords = Y[: , 0]
y_coords = Y[: , 1]
# display scatter plot
plt.scatter(x_coords, y_coords, s=1, c='black')
    
plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
plt.show()

with open('./constant/valid_E9.json') as fin:
    datastore = json.loads(fin.read())
    
z, sentences=datastore['z'] , datastore['target_sents'] 
# find tsne coords for 2 dimensions
tsne = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
Y = tsne.fit_transform(z)

plt.subplot(2,2,2)
plt.axis('off')
plt.title('Constant')
x_coords = Y[:2000, 0]
y_coords = Y[:2000, 1]
# display scatter plot
plt.scatter(x_coords, y_coords, s=1, c='black')
    
plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
plt.show()

with open('./linear/valid_E9.json') as fin:
    datastore = json.loads(fin.read())
    
z, sentences=datastore['z'][:2000], datastore['target_sents'][:2000]
# find tsne coords for 2 dimensions
tsne = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
Y = tsne.fit_transform(z)

plt.subplot(2,2,3)
plt.axis('off')
plt.title('Linear')
x_coords = Y[:2000, 0]
y_coords = Y[:2000, 1]
# display scatter plot
plt.scatter(x_coords, y_coords, s=1, c='black')
    
plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
plt.show()

with open('./cyclic/valid_E9.json') as fin:
    datastore = json.loads(fin.read())
    
z, sentences=datastore['z'][:2000], datastore['target_sents'][:2000]
# find tsne coords for 2 dimensions
tsne = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
Y = tsne.fit_transform(z)

plt.subplot(2,2,4)
plt.axis('off')
plt.title('Cyclical')
x_coords = Y[:2000, 0]
y_coords = Y[:2000, 1]
# display scatter plot
plt.scatter(x_coords, y_coords, s=1, c='black')
    
plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import json
from tqdm import tqdm
%matplotlib notebook

plt.figure()
 
for i in range(0,10):
    with open('./2020-Feb-16-14:27:26/valid_E{}.json'.format(i)) as fin:
        datastore = json.loads(fin.read())

    z, sentences=datastore['z'][:2000], datastore['target_sents'][:2000]
    # find tsne coords for 2 dimensions
    tsne = TSNE(n_components=2, random_state=0)
    np.set_printoptions(suppress=True)
    Y = tsne.fit_transform(z)

    plt.subplot(4,3,i+1)
    plt.axis('off')

    sched_type='cyclical-'+str(i)
    if i>4:
        sched_type='zero-'+str(i)
    plt.title(sched_type)
    x_coords = Y[: , 0]
    y_coords = Y[: , 1]
    # display scatter plot
    plt.scatter(x_coords, y_coords, s=1, c='black')

    plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
    plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
    plt.show()

```
